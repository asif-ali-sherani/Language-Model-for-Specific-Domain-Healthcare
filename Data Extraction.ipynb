# **Medical Question Answer Data Extraction**

### Training Dataset

```python
# Importing libraries
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
# training_dataset1_path = 'LiveQA_MedicalTask_TREC2017-master\TrainingDatasets\TREC-2017-LiveQA-Medical-Train-1.xml'
# training_dataset2_path = 'LiveQA_MedicalTask_TREC2017-master\TrainingDatasets\TREC-2017-LiveQA-Medical-Train-2.xml'
training_dataset1_path = '/content/drive/MyDrive/Colab Notebooks/Masters Thesis - Project/LiveQA_MedicalTask_TREC2017-master/TrainingDatasets/TREC-2017-LiveQA-Medical-Train-1.xml'
training_dataset2_path = '/content/drive/MyDrive/Colab Notebooks/Masters Thesis - Project/LiveQA_MedicalTask_TREC2017-master/TrainingDatasets/TREC-2017-LiveQA-Medical-Train-2.xml'
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Initialize lists to store question-answer pairs
question_list = []
answer_list = []
```

```python
# Parse XML Data
tree = ET.parse(training_dataset1_path)
root = tree.getroot()

# Extract question-answer pairs
for question_elem in root.findall('NLM-QUESTION'):
    question_text = question_elem.find('MESSAGE').text
    answers = [answer_elem.text for answer_elem in question_elem.findall('.//ANSWER')]

    # Concatenate different answers for the same question
    concatenated_answers = ', '.join(answers)

    # Append to lists
    question_list.append(question_text)
    answer_list.append(concatenated_answers)
```

```python
# Parse XML Data
tree = ET.parse(training_dataset2_path)  # Replace with your actual file path
root = tree.getroot()

# Extract question-answer pairs
for question_elem in root.findall('NLM-QUESTION'):
    question_text = question_elem.find('MESSAGE').text
    answers = [answer_elem.text for answer_elem in question_elem.findall('.//ANSWER')]

    # Concatenate different answers for the same question
    concatenated_answers = ', '.join(answers)

    # Append to lists
    question_list.append(question_text)
    answer_list.append(concatenated_answers)
```

```python
# Create DataFrame
training_dataset = pd.DataFrame({'Question': question_list, 'Answer': answer_list})
```

### Test Dataset

```python
test_data_path = '/content/drive/MyDrive/Colab Notebooks/Masters Thesis - Project/LiveQA_MedicalTask_TREC2017-master/TestDataset/TREC-2017-LiveQA-Medical-Test.xml'
```

```python
tree = ET.parse(test_data_path)  # Replace with your actual file path
root = tree.getroot()

# Initialize lists to store question-answer pairs
question_list = []
answer_list = []

# Extract question-answer pairs
for question_elem in root.findall('NLM-QUESTION'):
    question_text = question_elem.find('Original-Question/MESSAGE').text
    answers = [answer_elem.text for answer_elem in question_elem.findall('.//ANSWER')]

    # Concatenate different answers for the same question
    concatenated_answers = ', '.join(answers)

    # Append to lists
    question_list.append(question_text)
    answer_list.append(concatenated_answers)

# Create DataFrame
test_dataset = pd.DataFrame({'Question': question_list, 'Answers': answer_list})
```
